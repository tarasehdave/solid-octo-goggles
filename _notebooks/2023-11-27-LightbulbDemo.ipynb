```python
code:
{% assign BITS = 8 %}
<style>
    td {
        text-align: center;
        vertical-align: middle;
    }
</style>
<!-- Display -->
<table>
    <thead>
        <tr class="header" id="table">
            <th>Plus</th>
            <th>Binary</th>
            <th>Hexadecimal</th>
            <th>Decimal</th>
            <th>Minus</th>
        </tr>
    </thead>
    <tbody>
        <tr>
            <!-- Button to increment binary value -->
            <td><div class="button" id="add1" onclick="add(1)">+1</div></td>
            <!-- Display for binary, hexadecimal, and decimal values -->
            <td id="binary">00000000</td>
            <td id="hexadecimal">0</td>
            <td id="decimal">0</td>
            <!-- Button to decrement binary value -->
            <td><div class="button" id="sub1" onclick="add(-1)">-1</div></td>
        </tr>
    </tbody>
</table>
<!-- Light bulbs -->
{% assign bits = BITS | minus: 1 %}
<table>
    <thead>
        <tr>
            <!-- Create light bulb images and buttons for each bit -->
            {% for i in (0..bits) %}
            <th>
                <img id="bulb{{ i }}" src="{{site.baseurl}}/images/light-bulb-off.jpeg" alt="" width="40" height="Auto">
                <div class="button" id="butt{{ i }}" onclick="toggleBit({{ i }})">Turn on</div>
            </th>
            {% endfor %}
        </tr>
    </thead>
    <tbody>
        <tr>
            <!-- Create input fields to represent each bit's value -->
            {% for i in (0..bits) %}
            <td><input type='text' id="digit{{ i }}" value="0" size="1" readonly></td>
            {% endfor %}
        </tr>
    </tbody>
</table>
<script>
    // Constants
    const BITS = {{ BITS }};
    const MAX = 2 ** BITS - 1;
    const IMAGE_ON = "{{site.baseurl}}/images/lightbulb_on.png";
    const IMAGE_OFF = "{{site.baseurl}}/images/light-bulb-off.jpeg"
    // Function to get the binary value from input fields
    function getBits() {
        return Array.from({ length: BITS }, (_, i) => document.getElementById(`digit${i}`).value).join('');
    }
    // Function to update display with binary, hexadecimal, and decimal values
    function setConversions(binary) {
        document.getElementById('binary').innerHTML = binary;
        document.getElementById('hexadecimal').innerHTML = parseInt(binary, 2).toString(16);
        document.getElementById('decimal').innerHTML = parseInt(binary, 2).toString();
    }
    // Function to toggle the value of a specific bit
    function toggleBit(i) {
        const dig = document.getElementById(`digit${i}`);
        const image = document.getElementById(`bulb${i}`);
        const butt = document.getElementById(`butt${i}`);
        const newValue = dig.value === '1' ? '0' : '1';
        // Update input field, light bulb image, and button text
        dig.value = newValue;
        image.src = newValue === '1' ? IMAGE_ON : IMAGE_OFF;
        butt.innerHTML = newValue === '1' ? 'Turn off' : 'Turn on';
        // Update binary value and display
        const binary = getBits();
        setConversions(binary);
    }
    // Function to add or subtract from the binary value
    function add(n) {
        const binary = getBits();
        let decimal = parseInt(binary, 2);
        // Increment or decrement the decimal value
        decimal = (n > 0) ? (MAX === decimal ? 0 : decimal + n) : (0 === decimal ? MAX : decimal + n);
        // Convert the result back to binary
        const newBinary = decimal.toString(2).padStart(BITS, '0');
        // Update display with new binary value
        setConversions(newBinary);
        // Update input fields, light bulb images, and button text for each bit
        newBinary.split('').forEach((digit, i) => {
            const image = document.getElementById(`bulb${i}`);
            const butt = document.getElementById(`butt${i}`);
            document.getElementById(`digit${i}`).value = digit;
            image.src = digit === '1' ? IMAGE_ON : IMAGE_OFF;
            butt.innerHTML = digit === '1' ? 'Turn off' : 'Turn on';
        });
    }
</script>
```
